This is a test to see how much of a DnD 5e character builder can be built in Python with variables, functions, and IF/THEN statements. It will hopefully improve over time, but this is the first iteration with the simplest code.

Have fun!

```python
#Libraries we'll be using. Mostly randint for dice.
from random import randint

#Greetings to know that the code has actually started.
print("""Welcome to a DnD 5e character generator, in Python! 
Options are limited at this time, but this should be a way to help set up a character
with prompts, using the dice rolling method. No guarantees about power levels, or completeness.
Enjoy!""")

#Let's start with defining some of the things that we're going to be filling in.
Name = ""
Race = ""
Class = ""
Level = 0
HP = 0
Hit_Dice = 0
Profiency = 2
Str = 0
Str_Mod = int((Str - 10) / 2)
Str_Sav = False
Dex = 0
Dex_Mod = int((Dex - 10) / 2)
Dex_Sav = False
Con = 0
Con_Mod = int((Con - 10) / 2)
Con_Sav = False
Int = 0
Int_Mod = int((Int - 10) / 2)
Int_Save = False
Wis = 0
Wis_Mod = int((Wis - 10) / 2)
Wis_Sav = False
Cha = 0
Cha_Mod = int((Cha - 10) / 2)
Cha_Sav = False
Speed = 0
Sorry = "Sorry, that's not a recognized option at this time. Please try again."
#Whew, this was a lot. Won't do skills yet, just going to be working on the main 6 attributes.
#We'll also call the _Mod variables again later to make them recalculate.


#Ok, let's make something to get your race input.
def race():
    choice = input("""First choice you're going to make is your race. We have three races currently available:
    Dwarf, Elf, and Human
    Type in your choice here: """)
    if choice.lower() == 'dwarf':
        dwarf()
        return
    elif choice.lower() == 'elf':
        elf()
        return
    elif choice.lower() == 'human':
        human()
        return
    else:
        print(Sorry)
        race()

def dwarf():
    #Dwarves! Two sub-races to chose from, but Con bonus either way.
    global Con, Speed
    Con += 2
    Speed += 25
    choice_sub = input("""Dwarves! A hearty people! But do you want to be a Hill Dwarf or a Mountain Dwarf?
    Type Hill or Mountain here: """)
    if choice_sub.lower == 'hill':
        global Wis
        Wis += 1
        global HP
        HP += 1
        global Race
        Race = "Hill Dwarf"
        print("Sturdy choice")
        return
    elif choice_sub.lower == 'mountain': 
        global Str
        Str += 2
        Race = "Mountain Dwarf"
        print("Strong choice")
        return
    else:
        print(Sorry)
        dwarf()
        
def elf():
    #Elves! a mystical tree preople. No Drow here!
    global Race, Dex, Speed, Int, Wis
    Dex += 2
    Speed += 30
    choice_sub = input("""Elves! A graceful people! But do you want to be a High Elf or a Wood Elf?
    Type High or Wood here: """)
    if choice_sub.lower() == 'high':
        Int += 1
        Race = "High Elf"
        print("Smart Choice")
        return
    elif choice_sub.lower() == 'wood':
        Wis += 1
        Speed += 5
        Race = "Wood Elf"
        print("Wise choice")
        return
    else:
        print(Sorry)
        elf()

def human():
    #Humans! Most flexible race there is. Also popular.
    global Str, Dex, Con, Int, Wis, Cha, Speed
    Str += 1
    Dex += 1
    Con += 1
    Int += 1
    Wis += 1
    Cha += 1
    Speed += 30
    print("Humans! Flexible, ambitious, though fleeting. A well-rounded choice")
    return

#Class choice time. This is going to be lengthy with lots of If/Thens.
def job():
    choice_job = input("""Time to select your class! We have three options at the moment:
    Cleric, Fighter, Rogue
    Type in your choice here: """)
    if choice_job.lower == 'cleric':
        cleric()
        return
    elif choice_job.lower == 'fighter':
        fighter()
        return
    elif choice_job.lower == 'rogue':
        rogue()
        return
    else:
        print(Sorry)
        job()

#actual class functions are going to be pretty boring right now, since we're just concerned with a couple stats.
def cleric():
    global HP, Hit_Dice, Wis_Sav, Cha_Sav, Level
    HP += 8
    Hit_Dice = 8
    Wis_Sav = True
    Cha_Sav = True
    Level = 1
    print("Clerics are usually considered the healers of the group,\
    but their divine connection lends them to so much more")
    return

def fighter():
    global HP, Hit_Dice, Str_Sav, Con_Sav, Level
    HP += 10
    Hit_Dice = 10
    Str_Sav = True
    Con_Sav = True
    Level = 1
    print("Fighters are the usually the meat shields of the group,\
    but are masters of their weapons and control the flow of battle")
    return

def rogue():
    global HP, Hit_Dice, Dex_Sav, Int_Sav, Level
    HP += 8
    Hit_Dice = 8
    Dex_Sav = True
    Int_Sav = True
    Level = 1
    print("Rogues are usually the tricky kleptomaniacs of the group,\
    but their dexterity lends them tools for creative solutions to problems.")
    return

    
#Time to roll for attributes. Using the 4d6 minus the lowest number method of rolling stats.
def attribute_roll():
    attr = []
    for i in range(6):
        roll = [randint(1,6), randint(1,6), randint(1,6), randint(1,6)]
        roll_num = sum(roll) - min(roll)
        attr.append(roll_num)
    attr_sort = sorted(attr, reverse = True)
    print("Rolling for stats...")
    print("Your numbers are:")
    print(attr_sort)
    return attr_sort
    
def attribute_assign(attr_sort):
    #This is going to assign the attributes that were rolled in the previous function.
    global Str, Dex, Con, Int, Wis, Cha
    count = 6
    stat = ['str', 'dex', 'con', 'int', 'wis', 'cha']
    while count > 0:
        #Should later change this string to list the still available attributes, instead of printing the full list.
        assign = input("""Which stat do you want to assign your %d to?
        type Str, Dex, Con, Int, Wis, or Cha: """ % (attr_sort[0]))
        if assign.lower() == 'str' and assign.lower() in stat:
            Str += attr_sort[0]
            stat.remove('str')
            attr_sort = attr_sort[1:len(attr_sort)]
            count -= 1
        elif assign.lower() == 'dex' and assign.lower() in stat:
            Dex += attr_sort[0]
            stat.remove('dex')
            attr_sort = attr_sort[1:len(attr_sort)]
            count -= 1
        elif assign.lower() == 'con' and assign.lower() in stat:
            Con += attr_sort[0]
            stat.remove('con')
            attr_sort = attr_sort[1:len(attr_sort)]
            count -= 1
        elif assign.lower() == 'int' and assign.lower() in stat:
            Int += attr_sort[0]
            stat.remove('int')
            attr_sort = attr_sort[1:len(attr_sort)]
            count -= 1
        elif assign.lower() == 'wis' and assign.lower() in stat:
            Wis += attr_sort[0]
            stat.remove('wis')
            attr_sort = attr_sort[1:len(attr_sort)]
            count -= 1
        elif assign.lower() == 'cha' and assign.lower() in stat:
            Cha += attr_sort[0]
            stat.remove('cha')
            attr_sort = attr_sort[1:len(attr_sort)]
            count -= 1
        else:
            print(Sorry)
    return

#Now we deal with the twidly bits of calculating attribute modifiers and adding the Con Mod to level 1 HP.
def modifier_calc():
    global Str_Mod, Dex_Mod, Con_Mod, Int_Mod, Wis_Mod, Cha_Mod, HP
    Str_Mod = int((Str - 10) / 2)
    Dex_Mod = int((Dex - 10) / 2)
    Con_Mod = int((Con - 10) / 2)
    Int_Mod = int((Int - 10) / 2)
    Wis_Mod = int((Wis - 10) / 2)
    Cha_Mod = int((Cha - 10) / 2)
    HP += Con_Mod
            
#OK, let's kick everything into motion, and start calling functions
race()
job()
attribute_assign(attribute_roll())
modifier_calc()

print("""OK! Here's what your Character looks like so far!

%s is a %s %s at Level %d
They have %d HP, and %dd%d Hit Dice, with a Speed of %d feet per round
Their Strength is %d, with a +%d modifier
Their Dexterity is %d, with a +%d modifiier
Their Constitution is %d, with a +%d modifiier
Their Intelligence is %d, with a +%d modifiier
Their Wisdom is %d, with a +%d modifiier
Their Charisma is %d, with a +%d modifiier""" % (Name, Race, Class, Level, HP, Level, Hit_Dice, Speed, Str, Str_Mod, Dex, Dex_Mod, Con, Con_Mod, Int, Int_Mod, Wis, Wis_Mod, Cha, Cha_Mod))
```
